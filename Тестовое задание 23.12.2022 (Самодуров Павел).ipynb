### 1. Получены N независимых наблюдений случайной величины X: X1, X2, … , XN.
По какой формуле (и почему) вы оцените дисперсию X ?

Так как нам нужно посчитать дисперсию для конкретного X, мы считаем сумму квадратов отклонений X от каждого
значения X и делим полученное на N

$D = \frac{\sum_{i=1}^{n}(x - x_i)^2}{n}$


```python
X = [1, 4, 7, 9, 13, 21]


def variance(x, data):
    ''' Считаем дисперсию для конкретного X'''
    summ = 0
    for i in data:
        summ += (x - i) ** 2
    return f'Дисперсия {x} = {summ / len(data)}'

# Считаем дисперсию для X=7
print(variance(7, X))
```

### 2. Получены N независимых наблюдений случайной величины X: X1, X2, … , XN.
По какой формуле (и почему) вы оцените дисперсию случайной величины Y = X1 + X2+ … + XN ?


В данном случае мы будем считать дисперсию как сумму квадратов отклонений каждого X от среднего X и делим полученное на N. Таким образом мы считаем дисперсию по всем значениям в массиве.

$D = \frac{\sum_{i=1}^{n}(x_i - \frac{\sum_{i=1}^{n}x_i}{n})^2}{n}$

```python
def variance_all(data):
    ''' Считаем дисперсию для Y = X1 + X2+ … + XN'''
    summ = 0
    avr = sum(data) / len(data)
    for i in data:
        summ += (i - avr) ** 2
    return f'Дисперсия Y = {summ / len(data)}'

# Считаем дисперсию для Y = X1 + X2+ … + XN
print(variance_all(X))
```

```python
# Считаем дисперсию с помощью NumPy
import numpy as np

print('Дисперсия Y', np.var(X))
```

### 4. Игрок участвует в игре: бросают игральный кубик и если выпало число от 1 до 4, то ему платят рубль. Если выпало 5 или 6, то ему платят 0.5 рубля и еще раз бросают кубик по тем же правилам. Каков средний выигрыш игрока?

```python
# Всего исходов комбинаций при двух бросках
n = 6 ** 2
# Исходов с выигрышем 2 рублей
m2 = 4 ** 2
# Исходов с выигрышем 1 рубля
m1 = 2 ** 2
# Исходов с выигрышем 1,5 рублей
m1_5 = n - m2 - m1
print(f'''
Исходов с выигрышем 2 рублей {m2}
Исходов с выигрышем 1,5 рублей {m1_5}
Исходов с выигрышем 1 рубля {m1}
''')
```

```python
mean = (m2 * 2 + m1 * 1 + m1_5 * 1.5) / n
print('Средний выигрыш', mean)
```
